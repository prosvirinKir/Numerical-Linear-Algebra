_2._ (4 pts) Derive analytical expression for the gradient and hessian of the function $f$:

$$ R(x) = \frac{(Ax, x)}{(x, x)}, $$

where $A$ is a symmetric real matrix. Why the gradient of this function is important in NLA you will know in the lectures later.

$$
\nabla R(x) = \frac{\partial R(x)}{\partial x}=\frac{\frac{\partial}{\partial x}(A x, x) \cdot(x, x)-\frac{\partial}{\partial x}(x, x) \cdot(A x, x)}{(x, x)^{2}}
=\frac{2 A x(x, x)-2 x(A x, x)}{(x, x)^{2}}=\frac{2 A x}{(x, x)}-\frac{(A x, x)}{(x, x)^{2}} \cdot 2 x
=\frac{2 A x}{(x, x)}-\frac{2 x}{(x, x)} \cdot R(x)=\frac{2}{(x, x)}(A x-R(x) \cdot x)
$$

$$
\Delta R(x) = \frac{\partial^{2}}{\partial x^{2}} R(x)=\frac{\partial}{\partial x}\left(\frac{\partial R(x)}{\partial x}\right)=\frac{\partial}{\partial x}\left[\frac{2(A x-R(x) \cdot x)}{(x, x)}\right]
=(A x-R(x) \cdot x) \frac{\partial}{\partial x}\left(\frac{2}{(x, x)}\right)+\frac{2}{(x, x)} \frac{\partial}{\partial x}(A x-R(x) \cdot x)=\\
=(A x-R(x) \cdot x)\left(\frac{-2 x}{(x, x)^{2}}\right)+\frac{2}{(x, x)}\left(A-x \cdot \nabla R^{\top}(x)-R(x) \cdot I\right)=\\
=-\frac{2\nabla R(x)}{(x, x)} x^{\top}+\frac{2}{(x, x)}\left(A-x \cdot \nabla R^{\top}(x)-R(x) \cdot I\right)
=\frac{2}{(x, x)}\left(A-x \cdot \nabla R^{\top}(x)-R(x) \cdot I- \nabla R(x) \cdot x^{\top}\right)
$$

The complexity of the product of matrices $(n, m)$ and $(k, m)$ is $O (n \cdot m \cdot k)$. The analytical formula for calculating gradients is a sequential product of matrices of the form $(n, n)$ or $(n, k)$ or $(k, n)$ or $(k, k)$. 

Therefore, the complexity of the product will be $O (n ^ 2 \cdot k)$ or $O (n \cdot k ^ 2)$. Then the final complexity will be of the form $O (n ^ 2 \cdot k) + O (n \cdot k ^ 2) = O (n \cdot k \cdot \max (n, k))$.

Therefore, for constant K, we should see a quadratic dependence for N on the graph. And vice versa, for constant N we should see a quadratic dependence for K.

````python
import numpy as np
import jax.numpy as jnp
from jax import grad, jit
import time
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def F_U(U, D, V, X):
    return 0.5 * jnp.sum((X - jnp.dot(U, jnp.dot(D, V))) ** 2)

def F_V(V, D, U, X):
    return 0.5 * jnp.sum((X - jnp.dot(U, jnp.dot(D, V))) ** 2)

def dFdU(U, D, V, X):
    return (U @ D @ V - X) @ V.T @ D.T

def dFdV(V, D, U, X):
    return D.T @ U.T @ (U @ D @ V - X)
````

Let's check that our derivatives are the same in both approaches

````python
n = 1000
k = 500
````

````python
X, U, D, V = np.random.rand(n,n), np.random.rand(n,k), np.diag(np.random.rand(k)), np.random.rand(k,n)
````

````python
JAX_dFdU = jit(grad(F_U))
JAX_dFdV = jit(grad(F_V))
````

````python
print(f'Relative error: {np.mean((JAX_dFdU(U, D, V, X) - dFdU(U, D, V, X)) / dFdU(U, D, V, X))}')
print(f'Relative error: {np.mean((JAX_dFdV(V, D, U, X) - dFdV(V, D, U, X)) / dFdV(V, D, U, X))}')
````

We can see that our derivatives are the same, hince we did it right. 

Now let's go to compare analytical and automatic approaches for computing gradients depending on number of N and K

### Analytical and automatic computations for $\frac{\partial F(U, V)}{\partial U}$

````python
n_timing = []
jax_n_timing = []
for n in range(1000, 10000, 500):
    k = 1000
    X, U, D, V = np.random.rand(n,n), np.random.rand(n,k), np.diag(np.random.rand(k)), np.random.rand(k,n)

    start = time.time()
    res = dFdU(U, D, V, X)
    stop = time.time()
    n_timing.append(stop - start)

    start = time.time()
    res = JAX_dFdU(U, D, V, X)
    stop = time.time()
    jax_n_timing.append(stop - start)
````

````python
k_timing = []
jax_k_timing = []
for k in range(1000, 10000, 500):
    n = 1000
    X, U, D, V = np.random.rand(n,n), np.random.rand(n,k), np.diag(np.random.rand(k)), np.random.rand(k,n)

    start = time.time()
    res = dFdU(U, D, V, X)
    stop = time.time()
    k_timing.append(stop - start)

    start = time.time()
    res = JAX_dFdU(U, D, V, X)
    stop = time.time()
    jax_k_timing.append(stop - start)
````

**NOTE**: The first reason for such range for N and K is that it is a popular number of hidden neurones for neural networks. 

And the second reason is that we also want to estimate computational complexity of computing these gradients (in big-O notation), so we need quite large number of elements.

````python
x = list(range(1000, 10000, 500))
plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.title('Timing for different N')
plt.plot(x, n_timing, color='blue', label='analytical')
plt.plot(x, jax_n_timing, color='red', label='jax')
plt.ylabel(f"time", fontsize=12)
plt.xlabel(f"amount of N", fontsize=12)

plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.title('Timing for different K')
plt.plot(x, k_timing, color='blue', label='analytical')
plt.plot(x, jax_k_timing, color='red', label='jax')
plt.ylabel(f"time", fontsize=12)
plt.xlabel(f"amount of K", fontsize=12)

plt.legend()
plt.grid()

plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.title('Timing for different N')
plt.plot(np.log(x), np.log(n_timing), color='blue', label='analytical')
plt.plot(np.log(x), np.log(jax_n_timing), color='red', label='jax')
plt.ylabel(f"ln(time)", fontsize=12)
plt.xlabel(f"ln(amount of N)", fontsize=12)

plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.title('Timing for different K')
plt.plot(np.log(x), np.log(k_timing), color='blue', label='analytical')
plt.plot(np.log(x), np.log(jax_k_timing), color='red', label='jax')
plt.ylabel(f"ln(time)", fontsize=12)
plt.xlabel(f"ln(amount of K)", fontsize=12)
plt.legend()
plt.grid()
plt.show()
````

# Problem 3. Compression of the fully-connected layers in neural network with simple architecture (30 pts)

In this problem we consider the neural network that performs classification of the dataset of images. 
Any neural network can be considered as composition of simple linear and non-linear functions.
For example, a neural network with 3 layers can be represented as 

$$f_3(f_2(f_1(x, w_1), w_2), w_3),$$

where $x$ is input data (in our case it will be images) and $w_i, \; i =1,\dots,3$ are parameters that are going to be trained. 

We will study the compression potential of neural network with simple architecture: alternating some numbers of linear and non-linear functions. 

The main task in this problem is to study how the compression of fully-connected layers affects the test accuracy.
Any fully-connected layer is represented as linear function $AX + B$, where $X$ is input matrix and $A, B$ are trainable matrices. Matrices $A$ in every layer are going to be compressed.
The main result that you should get is the plot of dependence of test accuracy on the total number of parameters in the neural network.

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class ResidualBlock(nn.Module):
    def __init__(self, in_features, middle_features, activation=nn.ReLU()):
        super().__init__()
        self.layer1 = nn.Linear(in_features, middle_features)
        self.layer2 = nn.Linear(middle_features, in_features)
        self.activation = activation
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.res1 = ResidualBlock(1000,1000)
        self.res2 = ResidualBlock(1000,1000)
        self.fc2 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.train()
    model.to(device)
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````
